In fact, the `=` operator, which you used in the previous section to create variables, is also called the *assignment* operator, and is a symbol used to create variables.

Below, we will explore some of the different kinds of operators and how they are used in Python code.

Arithmetic operators are symbols used in math equations. They include the `+`, `-`, `*`, and `/` symbols, which are used to perform addition, subtraction, multiplication, and division, respectively. These symbols can be used between two numbers in any code to give the result of the equation.

## **Conditional Filtering**
#### **Comparison Operators**

Comparison operators are used to compare two values and return a boolean result (**True** or **False**). These operators include `==`, `!=`, `>`, `<`, `>=`, and `<=`. For example, `==`checks if two values are equal, `!=` checks if they are not equal, `>` checks if the left value is greater than the right value, `<` checks if the left value is less than the right value, `>=` checks if the left value is greater than or equal to the right value, and `<=` checks if the left value is less than or equal to the right value. These operators are essential for making decisions in code, such as in conditional statements and loops.

#### **Conditions**
In pandas we can create multiple conditions and negate conditions using the symbols `&`, `|`, and `~`.

| If statement | pandas | Pronunciation | Meaning                                                                                                                                                |
|--------------|--------|---------------|--------------------------------------------------------------------------------------------------------------------------------------------------------|
| `and`        | `&`    | "ampersand"   | `&` means that both criteria must be `True` in order for the entire condition to be `True`.                                                            |
| `or`         | `\|`   | "pipe"        | `\|` means that just one of the criteria needs to be `True` in order for the entire condition to be `True`.                                            |
| `not`        | `~`    | "tilda"       | The `~` symbol is placed in front of a condition in pandas to negate the condition (do the opposite). If it would have been `True`, it is now `False`. |

To use multiple conditions for a filter, each condition should be enclosed in its own set of parentheses, and the `&` or `|` symbols placed between them.

```python
# Filter the data frame using operators and conditions like ==, <=, <, >, >=, ~, != and &, |, ~
df['pokename'] == 'Natu'
```

```python
# Use the != or ~ operator
df[~(df['pokename'] == 'Natu')]

df[(df['pokename'] != 'Natu')]
```

```python
# Use the | (or) operator
df[(df['pokename'] == 'Natu') | (df['pokename'] == 'Vileplume')]
```

✅ Try it out: Filter the Pokemon dataframe by selecting either the Pokemon named 'Natu' or 'Vileplume' and only return the rows where the Pokemon placed between 1st and 3rd.

```python
df[((df['pokename'] == 'Natu') | (df['pokename'] == 'Vileplume')) & (df['place'] <= 3) & (df['place'] > 0 )]
```
