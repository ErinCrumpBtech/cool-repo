```python
# Import statement
import pandas as pd

# Load data
df = pd.read_csv('horse_racing.csv')
df.head()
```

```python
# Question 1: Of all the rows in the data set what was the average weight of a jockey? Weights are listed in kilograms. Round to two decimal places.
round(df['Jockey weight'].mean(),2)
```

```python
# Question 2: What was the average race time for jockey Brett Prebble? Round to two decimal places.
df_bp = df[df['Jockey'] == 'Brett Prebble']
round(df_bp['Race time'].mean(), 2)
```

```python
# Question 3: Create a new dataframe that contains rows where horses and their jockeys won first place. How much money did first place jockeys win, on average?
# The money unit is Hong Kong dollars (HKD). Round to two decimal places.
df_first = df[df['Final place'] == 1]
round(df_first['Prize money'].mean(),2)
```

```python
# Question 4: Did winning horses tend to run on dirt or grass surfaces?
df_first['Surface'].value_counts(normalize=True)
```

```python
# Question 5: What was the fastest race time among horses that ran the 1400 meter race? The units given are in seconds
# (sometimes, the data is dirty, just assume they are all seconds for now).
df_1400 = df[df['Distance'] == 1400]
df_1400['Race time'].min()
```

```python
df_1400['Country'].value_counts()
```

```python
# Question 6: What was the slowest race time among Irish horses who ran the 1400 meter race?
df_1400[df_1400['Country'] == 'Irland'].sort_values(by = 'Race time', ascending= False).head(1)
```

```python
# Question 7: Which Australian jockey has competed in the most races?
df_aussie = df[df['Country'] == 'Australia']
df_aussie['Jockey'].value_counts().head(1)
```

```python
# Question 8: Assume that greater odds indicates a lesser likelihood to win the race. According to this data set, which jockey out of Australia, Ireland, or the New Zealand
# has the least likelihood to win a race, on average (ie. the highest odds)?
df_mc = df[df['Country'].isin(['Australia', 'Irland', 'New Zealand'])]

df_mc[['Jockey','Odds']].groupby('Jockey').mean().sort_values(by='Odds', ascending = False).head(1)
```

```python
# Question 9: Of all horses age 10 or above, what is the horse ID of the one that has the best average race time (ie. the smallest)?
df_10 = df[df['Horse age'] >= 10]
df_10[['HorseId', 'Race time']].groupby('HorseId').mean().sort_values(by = 'Race time', ascending = True).head(1)
```

```python
# Question 10: Which country won the most races (first place) where the prize money was over $2,000,000?
df_price = df[(df['Prize money'] > 2000000) & (df['Final place'] == 1)]
df_price['Country'].value_counts().head(1)
```
